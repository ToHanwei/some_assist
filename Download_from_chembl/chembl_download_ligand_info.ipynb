```python
target = new_client.target
activity = new_client.activity
gene_name = 'ADORA2A'

target_df = pd.DataFrame.from_dict(target_dict)
target_homo = target_df[target_df['tax_id']==9606]

all_df = pd.DataFrame()
for target_id in target_homo["target_chembl_id"]:
    value = activity.filter(target_chembl_id=target_id)
    compud_df = pd.DataFrame.from_dict(value)
    all_df = all_df.append(compud_df)
    
print(all_df)
```

```python
pki = all_df[all_df['type']=='pKi']
ki = all_df[all_df['type']=='Ki']
pec50 = all_df[all_df['type']=='pEC50']
ec50 = all_df[all_df['type']=='EC50']
```

```python
all_ki = pki.append(ki)
all_ec50 = pec50.append(ec50)
```

```python
merge_df = all_ki.merge(all_ec50, left_on=['target_chembl_id', 'molecule_chembl_id'], right_on=['target_chembl_id', 'molecule_chembl_id'], suffixes=("_Ki", "_EC50"))
```

```python
ki_ec50 = merge_df[['target_chembl_id', 'molecule_chembl_id', 'standard_value_Ki', 'standard_units_Ki',  'standard_value_EC50', 'standard_units_EC50', 'document_chembl_id_Ki', 'document_chembl_id_EC50']]
```
