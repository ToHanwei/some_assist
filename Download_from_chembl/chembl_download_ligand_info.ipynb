```python
target = new_client.target
activity = new_client.activity
gene_name = 'ADORA2A'
target_dict = target.filter(target_synonym__icontains=gene_name)
```

```python
target_df = pd.DataFrame.from_dict(target_dict)
```

```python
target_homo = target_df[target_df['tax_id']==9606]
```

```python
target_homo["target_chembl_id"]
```

```python
result_df = pd.DataFrame()
for target_id in target_homo["target_chembl_id"]:
    value = activity.filter(target_chembl_id=target_id, assay_type="B").filter(standard_type="Ki")
    compud_df = pd.DataFrame.from_dict(value)
    result_df = result_df.append(compud_df)
    
print(result_df)
```
